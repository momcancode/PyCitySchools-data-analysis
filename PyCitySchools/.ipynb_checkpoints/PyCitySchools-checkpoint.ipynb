## PyCitySchools Data Overview

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=[
                                "school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

```python
passing_score = 70
math_passing_df = school_data_complete[school_data_complete["math_score"] >= passing_score]
reading_passing_df = school_data_complete[school_data_complete["reading_score"] >= passing_score]
overall_passing_df = school_data_complete[(school_data_complete["math_score"] >= passing_score) &
                                          (school_data_complete["reading_score"] >= passing_score)]
```

```python
def district_summary():
    summary = dict()

    summary["Total Schools"] = school_data["school_name"].count()

    total_student = student_data["student_name"].count()
    summary["Total Students"] = total_student

    summary["Total Budget"] = school_data["budget"].sum()

    summary["Average Math Score"] = school_data_complete["math_score"].mean()
    summary["Average Reading Score"] = school_data_complete["reading_score"].mean()

    summary["% Passing Math"] = len(math_passing_df)/total_student*100
    summary["% Passing Reading"] = len(reading_passing_df)/total_student*100
    summary["% Overall Passing"] = len(overall_passing_df)/total_student*100

    return pd.DataFrame([summary])
```

```python
district_summary = district_summary()
```

```python
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map(
    "{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map(
    "${:,.2f}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
def school_summary_partial():
    school_groupby = school_data_complete.groupby("school_name")

    school_type = school_groupby["type"].last()

    num_student_per_school = school_groupby["student_name"].count()
    school_budget = school_groupby["budget"].last()
    budget_per_student = school_budget/num_student_per_school

    mathscore_average = school_groupby["math_score"].mean()
    reading_average = school_groupby["reading_score"].mean()

    math_passing_num_per_school = school_groupby["math_score"].apply(
        lambda x: x[x >= passing_score].count())
    math_passing_percent = math_passing_num_per_school/num_student_per_school*100

    reading_passing_num_per_school = school_groupby["reading_score"].apply(
        lambda x: x[x >= passing_score].count())
    reading_passing_percent = reading_passing_num_per_school/num_student_per_school*100

    return pd.DataFrame({"School Type": school_type,
                         "Total Students": num_student_per_school,
                         "Total School Budget": school_budget,
                         "Per Student Budget": budget_per_student,
                         "Average Math Score": mathscore_average,
                         "Average Reading Score": reading_average,
                         "% Passing Math": math_passing_percent,
                         "% Passing Reading": reading_passing_percent
                         })
```

```python
school_summary = school_summary_partial()
```

```python
overall_passing_num_per_school = overall_passing_df.groupby(
    "school_name").size().rename("overall_passing_num")
overall_passing_num_per_school
```

```python
merge_df = pd.merge(
    school_summary, overall_passing_num_per_school, on="school_name", how="left")
merge_df["% Overall Passing"] = merge_df["overall_passing_num"] / merge_df["Total Students"]*100

del merge_df["overall_passing_num"]

merge_df["Total School Budget"] = merge_df["Total School Budget"].map("${:,.2f}".format)
merge_df["Per Student Budget"] = merge_df["Per Student Budget"].map("${:,.2f}".format)
merge_df
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

```python
def scores_by_grade(df, values, index="school_name", columns="grade", aggfunc=np.mean):
    scores_by_grade_df = pd.pivot_table(
        df, values=values, index=index, columns=columns, aggfunc=aggfunc)
    return scores_by_grade_df[["9th", "10th", "11th", "12th"]]
```

```python
scores_by_grade(school_data_complete, values="math_score")
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
scores_by_grade(school_data_complete, values="reading_score")
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate

```python
def school_performance(df, current_column, new_column=None, bins=None, labels=None, is_binning=True):
    resetdf = df.reset_index()
    selected_columns = ["Average Math Score", "Average Reading Score",
                        "% Passing Math", "% Passing Reading", "% Overall Passing"]
    if is_binning == True:
        resetdf[new_column] = pd.cut(
            resetdf[current_column], bins=bins, labels=labels, include_lowest=True)
        outcome_df = resetdf.groupby(
            new_column)[selected_columns].mean().round(2)
    else:
        outcome_df = resetdf.groupby(current_column)[
            selected_columns].mean().round(2)
    return outcome_df
```

```python
spending_bins = [0, 584, 629, 644, 675]
spending_labels = ["<$584", "$585-629", "$630-644", "$645-675"]

school_performance(merge_df, current_column="Per Student Budget",
                   new_column="Spending Ranges (Per Student)", bins=spending_bins, labels=spending_labels)
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_performance(merge_df, current_column="Total Students", new_column="School Size",
                   bins=size_bins, labels=size_labels)
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
school_performance(merge_df, current_column="School Type", is_binning=False)
```

## Observable Trends based on the data

* The top five performing schools are all of the charter types. Meanwhile, the bottom five performing schools are all of the district types.

* In general, the average math and reading scores over different grade levels at each school are not significantly different. The question is whether a student's academic foundation starting high school is the main contributor to that school performance over time.

* Higher funding per student does not necessarily guarantee higher overall academic performance. Notably, the higher the spending ranges per student, the lower the overall passing percentage.

* Schools of large size (from 2000 to 5000 students) are the worst performers with lowest overall passing percentage, especially significantly lower average math performance compared to small and medium school sizes.

* Charter schools have significantly higher overall passing percentage than district schools.
